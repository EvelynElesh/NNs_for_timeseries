# Ploting the model results

```python


fig = plt.figure(figsize=(25,8));

ax = fig.add_subplot(111);
ax.plot(All_join_val['date'], All_join_val['LSTM'], color = (0,0,0), linewidth = 4, alpha = .9, label = 'LSTM');
ax.plot(All_join_val['date'], All_join_val['Fp'], color = (1,0,0), label = 'FbProphet');
ax.plot(All_join_val['date'], All_join_val['NeuralProphet'], color = (0,1,1), label = 'NeuralProphet')
ax.plot(All_join_val['date'], All_join_val['SARIMA'], color = (1,0,1), label = 'SARIMA')
ax.plot(All_join_val['date'], All_join_val['ARIMA'], color = (0,1,0), label = 'ARIMA')
ax.plot(All_join_val['date'], All_join_val['Actual'], color = (0,0,1), label = 'Actual')
ax.set_title('Comparing different models')
ax.set_xlabel('Sales')
ax.set_ylabel('Date')
ax.legend(loc='lower right')
plt.xticks(rotation='vertical');
```
